```python
#spilling the datasets in x and y
y=df['bwt']
x=df.iloc[:,1:]
```

```python
x.head()
```

```python
#Now making the ML model train-test-split

from sklearn.model_selection import train_test_split as tts

x_test,x_train,y_test,y_train=tts(x,y,test_size=.20)

from sklearn.linear_model import LinearRegression,Lasso, Ridge

lr=LinearRegression()
lass=Lasso()
rid=Ridge()


lr.fit(x_train, y_train)
lass.fit(x_train,y_train)
rid.fit(x_train,y_train)

#Now put the prediction in here!

lr_predict=lr.predict(x_test)
lass_predict=lass.predict(x_test)
rid_predict=rid.predict(x_test)

#Now to check the prediction with the real values!


from sklearn.metrics import r2_score, mean_squared_error

lr_r=r2_score(y_test,lr_predict)
lass_r=r2_score(y_test,lass_predict)
rid_r=r2_score(y_test,rid_predict)



lr_mse=mean_squared_error(y_test,lr_predict)
lass_mse=mean_squared_error(y_test,lass_predict)
rid_mse=mean_squared_error(y_test,rid_predict)


print("\nPerformnace of the model evaluation Linear regression")
print("The r2_square value:",lr_r)
print("The mean square value is:",lr_mse)


print("\nPerformnace of the model evaluation Lasso regression")
print("The r2_square value:",lass_r)
print("The mean square value is:",lass_mse)


print("\nPerformnace of the model evaluation ridge regression")
print("The r2_square value:",rid_r)
print("The mean square value is:",rid_mse)


import pickle
with open ('model/model.pkl',"wb") as f:
    pickle.dump(lr,f)

```
